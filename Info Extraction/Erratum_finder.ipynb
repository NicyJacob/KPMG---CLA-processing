```python
def pdf2info(pdf,search_keyword):
    ''' Function to convert pdf to text and find matching sentences'''
    try:
        
        #Now give the pdf name
        pdf = pdf
        pdfFileObj = open(pdf, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        #print(pdfReader.numPages) # will give total number of pages in pdf
        text_extract = []
        for i in range(pdfReader.numPages):
            pageObj = pdfReader.getPage(i)
            text=(pageObj.extractText())
            text=text.split(",")
            text_extract.append(text)
            
        x = erratum_finder(text_extract,search_keyword)
        return(x)
    except IOError:
        pass
```

```python
def erratum_finder(text_extract, search_keyword):
    '''Function to extract the ERRATUM with the matching search_keywords'''
    text = text_extract
    #print(text)
    #words = search_keywords
    for txt in text:
        for t in txt:
            txt1 = re.sub(r"\n+", " ", t)
            #print(txt1)
            sentences=sent_tokenize(txt1)
            #print(sentences)
            for sentence in sentences:
                words = word_tokenize(sentence.lower())
                for word in words:
                    if word == search_keyword.lower():
                        x = text[0]
                        x1 = str(x)
                        y = re.sub(r"\n+", " ", x1)
                        return y
```

```python
def get_prev_cla(extracted_sentences):
    extractedsentences = str(extracted_sentences)
    #print(type(extractedsentences))

    sentences=sent_tokenize(extractedsentences)
    #print(sentences)
    for sentence in sentences:
        words = word_tokenize(sentence.lower())
        #print(words)
        for word in words:
            if word.startswith("nr."):
                x = word
                return(x)
```
