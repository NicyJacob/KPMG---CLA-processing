```python
def pdf2info(pdf,search_keyword):
    ''' Function to convert pdf to text and find matching sentences'''
    try:
        
        #Now give the pdf name
        pdf = pdf
        pdfFileObj = open(pdf, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        #print(pdfReader.numPages) # will give total number of pages in pdf
        text_extract = []
        for i in range(pdfReader.numPages):
            pageObj = pdfReader.getPage(i)
            text=(pageObj.extractText())
            text=text.split(",")
            text_extract.append(text)
            #for paragraph in text:
                #lst = []
                #for word in search_keywords:
                    #if word in paragraph:
                        #frequency +=1
                        #lst.append(frequency)
                        #sentences = re.sub(r"\n+", " ", paragraph)
                        #sentences = re.findall(r"([^.]*?vantoepassing[^.]*\.)",sentences)
                        #text_extract.append(sentences)
        #return(text_extract)
        x = sentence_finder(text_extract,search_keyword)
        return(x)
    except IOError:
        pass
    
```

```python
def sentence_finder(text_extract, search_keyword):
    '''Function to extract the sentence with the matching search_keywords'''
    text = text_extract
    #print(text)
    #words = search_keywords
    for txt in text:
        for t in txt:
            txt1 = re.sub(r"\n+", " ", t)
            #print(txt1)
            sentences=sent_tokenize(txt1)
            #print(sentences)
            for sentence in sentences:
                words = word_tokenize(sentence.lower())
                for word in words:
                    if word == search_keyword.lower():
                        x = sentence
                        return x
        
```

```python
## MAIN ##

import PyPDF2
import pandas as pd
import re
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

root_data_path = "C:/Users/nicyj/becode/KPMG"
folder = "/Data/NL_split_pdfs/"
#pdf_filename = "100-2021-015127.pdf"

data = pd.read_excel(root_data_path + '/Data/CLA_meta_from_2018_-_Translated_commitee_details.xlsx',engine='openpyxl')
pdf_list = data['NL_split_pdf_files']
erratum = []
vantoepassing = []
eindejaarspremie = []
geldigheidsduur = []
for i in range(len(pdf_list)):
    pdf_filename = pdf_list[i]
    #print(pdf_filename)
    pdf = root_data_path + folder + pdf_filename
    
    #search_keywords=['ERRATUM','vantoepassing','eindejaarspremie','geldigheidsduur']
    
    search_keywords=['ERRATUM']
    for search_keyword in search_keywords:
        extracted_sentences = pdf2info(pdf, search_keyword)
        #print(extracted_sentences)
        erratum.append(extracted_sentences)
        

    search_keywords=['vantoepassing']
    for search_keyword in search_keywords:
        extracted_sentences = pdf2info(pdf, search_keyword)
        vantoepassing.append(extracted_sentences)
        
    search_keywords=['eindejaarspremie']
    for search_keyword in search_keywords:
        extracted_sentences = pdf2info(pdf, search_keyword)
        eindejaarspremie.append(extracted_sentences)
        
    search_keywords=['geldigheidsduur']
    for search_keyword in search_keywords:
        extracted_sentences = pdf2info(pdf, search_keyword)
        geldigheidsduur.append(extracted_sentences)
        
print(len(erratum))
print(len(vantoepassing))
data['ERRATUM'] = erratum
data['vantoepassing'] = vantoepassing
data['eindejaarspremie'] = eindejaarspremie
data['geldigheidsduur'] = geldigheidsduur

data.head()
```
